## Importing Libraries

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
comparison_df = pd.DataFrame({

    "Model": [
        "Logistic Regression",
        "Random Forest",
        "DistilBERT",
        "RoBERTa"
    ],

    "Accuracy": [
        0.9998,
        0.9999,
        0.9976,
        0.9982
    ],

    "Precision": [
        0.9998,
        0.9999,
        0.9976,
        0.9982
    ],

    "Recall": [
        0.9998,
        0.9999,
        0.9976,
        0.9982
    ],

    "F1 Score": [
        0.9998,
        0.9999,
        0.9976,
        0.9982
    ],

    "Training Time (min)": [
        3,
        8,
        70,
        120
    ],

    "Inference Speed": [
        95,
        80,
        35,
        28
    ]
})

print(comparison_df)
```

## Accuracy Comparison

```python
plt.figure(figsize=(10,6))

sns.barplot(
    x="Model",
    y="Accuracy",
    data=comparison_df
)

plt.title("Model Accuracy Comparison")

plt.xticks(rotation=15)

plt.savefig(
    r"C:\Users\G VARSHIT HARIPRASAD\Desktop\Infotact Solutions Project\Month 2 - April\AI Sentiment Analysis System 2\outputs\model_accuracy_comparison.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## Precision / Recall / F1 Comparison

```python
metrics_df = comparison_df.melt(
    id_vars="Model",
    value_vars=[
        "Precision",
        "Recall",
        "F1 Score"
    ]
)

plt.figure(figsize=(12,6))

sns.barplot(
    data=metrics_df,
    x="Model",
    y="value",
    hue="variable"
)

plt.title("Precision, Recall & F1 Score Comparison")

plt.xticks(rotation=15)

plt.savefig(
    r"C:\Users\G VARSHIT HARIPRASAD\Desktop\Infotact Solutions Project\Month 2 - April\AI Sentiment Analysis System 2\outputs\precision_recall_f1_comparison.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## Training Time Comparison

```python
plt.figure(figsize=(10,6))

sns.barplot(
    x="Model",
    y="Training Time (min)",
    data=comparison_df
)

plt.title("Training Time Comparison")

plt.xticks(rotation=15)

plt.savefig(
    r"C:\Users\G VARSHIT HARIPRASAD\Desktop\Infotact Solutions Project\Month 2 - April\AI Sentiment Analysis System 2\outputs\training_time_comparison.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## Radar Graph

```python
import numpy as np

categories = [
    "Accuracy",
    "Precision",
    "Recall",
    "F1 Score"
]

N = len(categories)

angles = np.linspace(
    0,
    2 * np.pi,
    N,
    endpoint=False
).tolist()

angles += angles[:1]

fig, ax = plt.subplots(
    figsize=(8,8),
    subplot_kw=dict(polar=True)
)

for i, row in comparison_df.iterrows():

    values = [
        row["Accuracy"],
        row["Precision"],
        row["Recall"],
        row["F1 Score"]
    ]

    values += values[:1]

    ax.plot(
        angles,
        values,
        label=row["Model"]
    )

ax.set_xticks(angles[:-1])

ax.set_xticklabels(categories)

plt.title("Model Performance Radar Chart")

plt.legend(
    loc="upper right",
    bbox_to_anchor=(1.3, 1.1)
)

plt.savefig(
    r"C:\Users\G VARSHIT HARIPRASAD\Desktop\Infotact Solutions Project\Month 2 - April\AI Sentiment Analysis System 2\outputs/model_radar_chart.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## Final Ranking Plot

```python
comparison_df = comparison_df.sort_values(
    by="Accuracy",
    ascending=False
)

plt.figure(figsize=(10,6))

sns.barplot(
    x="Accuracy",
    y="Model",
    data=comparison_df
)

plt.title("Final Model Ranking")

plt.savefig(
    r"C:\Users\G VARSHIT HARIPRASAD\Desktop\Infotact Solutions Project\Month 2 - April\AI Sentiment Analysis System 2/outputs/final_model_ranking.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
