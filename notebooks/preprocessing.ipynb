```python
df = pd.read_csv(r"C:\Users\G VARSHIT HARIPRASAD\Desktop\Infotact Solutions Project\Month 2 - April\AI Sentiment Analysis System 2\data/raw_dataset.csv")
```

```python
df["text"] = (

    df["Problem (formerly Complaint Type)"].astype(str)

    + " " +

    df["Problem Detail (formerly Descriptor)"].astype(str)

    + " " +

    df["Additional Details"].astype(str)
)
```

```python
def clean_text(text):

    text = text.lower()

    text = re.sub(r"http\\S+", "", text)

    text = re.sub(r"[^a-zA-Z\\s]", "", text)

    text = re.sub(r"\\s+", " ", text).strip()

    return text
```

```python
df["clean_text"] = df["text"].apply(clean_text)
```
